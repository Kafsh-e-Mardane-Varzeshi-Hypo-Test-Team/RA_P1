```python
import pandas as pd
import os
import math
```

```python
path = "F:\\Regression Analysis\\RA_P1\\data\\traffic\\nowruz_03\\"
pathToSave = "F:\\Regression Analysis\\RA_P1\\data\\avg_speed_sorted\\"
```

```python
files = os.listdir(path)
dfs = [pd.read_csv(path + file, header=0) for file in files]
```

```python
def get_v(df):
    sum_v = 0
    count_v = 0
    sum_total_v = 0
    count_total_v = 0
    for index, row in df.iterrows():
        if not math.isnan(row.avg_speed):
            if (str(row.date) >= "1403-01-01" and (str(row.date) <= "1403-01-03")):
                sum_v += row.avg_speed
                count_v += 1
            sum_total_v += row.avg_speed
            count_total_v += 1

    if (count_v == 0):
        return -1, -1
    mean_v = sum_v/count_v
    mean_total_v = sum_total_v/count_total_v
    print(f"code: {df.code[0]}, sum: {sum_v}, count: {count_v}, mean: {mean_v}")
    print(f"code: {df.code[0]}, sum: {sum_total_v}, count: {count_total_v}, mean: {mean_total_v}")
    print()
    return int(mean_v), int(mean_total_v)
```

```python
v_sorted = {}
for i in range(len(dfs)):
    df = dfs[i]
    df_v, df_total_v = get_v(df)
    if df_v != -1:
        v_sorted[i + 1] = {'code': df.code[0].astype(str), 'name': df.name[0], 'avg_speed_1to3': df_v, 'avg_spped_total': df_total_v, 'diff_speed': df_total_v-df_v}

df = pd.DataFrame.from_dict(v_sorted, orient='index', columns=['code', 'name', 'avg_speed_1to3', 'avg_spped_total', 'diff_speed'])
df = df.sort_values(by="diff_speed", ascending = False)
df.to_csv(pathToSave + "avg_speed_sorted_03.csv", index=False)
df
```
